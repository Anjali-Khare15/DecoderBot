```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)  # Set 'Date' column to index

# Split dates
split_date = pd.to_datetime('2020-01-01')
eval_date = pd.to_datetime('2022-01-01')

# Split
data_train = df.loc[df.index <= split_date].copy()
data_test = df.loc[(df.index > split_date) & (df.index <= eval_date)].copy()
data_eval = df.loc[df.index > eval_date].copy()
```

```python
data_eval.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

#'Open' is feature we want to predict
feature_col = 'Open'
target_col = 'Open'

# Scale feature data
scaler = MinMaxScaler()
data_train_scaled = scaler.fit_transform(data_train[[feature_col]])
data_test_scaled = scaler.transform(data_test[[feature_col]])
data_eval_scaled = scaler.transform(data_eval[[feature_col]])

# sequences generator
def create_sequences(data, sequence_length):
    xs = []
    ys = []
    for i in range(len(data) - sequence_length):
        x = data[i:(i + sequence_length)]
        y = data[i + sequence_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

sequence_length = 60  # Number of days to look back to predict the next day

# Create sequences
X_train, y_train = create_sequences(data_train_scaled, sequence_length)
X_test, y_test = create_sequences(data_test_scaled, sequence_length)
X_eval, y_eval = create_sequences(data_eval_scaled, sequence_length)
```
